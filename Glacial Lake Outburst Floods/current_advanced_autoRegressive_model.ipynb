```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python

OUT_STEPS = 24
def create_dataset(data, time_steps=OUT_STEPS):
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:i+time_steps])
        y.append(data[i+1:i+1+time_steps])
    return np.array(X), np.array(y)

train_data, val_data = train_test_split(data, test_size=0.2, shuffle=False)

train_X, train_y = create_dataset(train_data)
val_X, val_y = create_dataset(val_data)
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
class FeedBack(tf.keras.Model):
    def __init__(self, units, out_steps, num_features):
        super().__init__()
        self.out_steps = out_steps
        self.units = units
        self.lstm_cell = tf.keras.layers.LSTMCell(units)
        self.lstm_rnn = tf.keras.layers.RNN(self.lstm_cell, return_state=True)

        self.lstm_cell_2 = tf.keras.layers.LSTMCell(units)
        self.lstm_rnn_2 = tf.keras.layers.RNN(self.lstm_cell_2, return_state=True, return_sequences=True)

        self.dense = tf.keras.layers.Dense(num_features)

    def warmup(self, inputs):
        x, *state = self.lstm_rnn(inputs)
        x = tf.expand_dims(x, axis=1)
        x, *state_2 = self.lstm_rnn_2(x)
        x = tf.squeeze(x, axis=1)
        prediction = self.dense(x)
        return prediction, state, state_2 

    def call(self, inputs, training=None):
        predictions = []
        prediction, state, state_2 = self.warmup(inputs) 
        predictions.append(prediction)

        for n in range(1, self.out_steps):
            x = prediction
            x, state = self.lstm_cell(x, states=state, training=training)
            x, state_2 = self.lstm_cell_2(x, states=state_2, training=training)
            prediction = self.dense(x)
            predictions.append(prediction)

        predictions = tf.stack(predictions)
        predictions = tf.transpose(predictions, [1, 0, 2])
        return predictions


units = 1024  
num_features = df.shape[1]
feedback_model = FeedBack(units=units, out_steps=OUT_STEPS, num_features=num_features)

feedback_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss=tf.keras.losses.MeanSquaredError(),
    metrics=[tf.keras.metrics.MeanAbsoluteError()],
    run_eagerly=True
)
```

```python
from tensorflow.keras.models import load_model

# Path to your H5 file
model_path = 'feedback_model.h5'

# Load the model
model = load_model(model_path)

# Check the model summary
model.summary()
```

```python
from tensorflow.keras.utils import plot_model

# Create a dummy input shape for visualization (batch_size, time_steps, features)
input_shape = (None, 20, df.shape[1])
feedback_model.build(input_shape)

# Visualize and save the model
plot_model(feedback_model, to_file="feedback_model.png", show_shapes=True, expand_nested=True)
```

```python
history = feedback_model.fit(train_X, train_y, epochs=50, validation_data=(val_X, val_y), batch_size=32)
```

```python
import matplotlib.pyplot as plt

# Plot training & validation loss values
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss', color='blue')
plt.plot(history.history['val_loss'], label='Validation Loss', color='orange')
plt.title('Model Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Evaluate the model on validation data
val_loss, val_mae = feedback_model.evaluate(val_X, val_y)
print(f'Validation Loss: {val_loss}, Validation MAE: {val_mae}')

# Plot some predictions
def plot_predictions(model, X, y):
    predictions = model.predict(X)
    plt.figure(figsize=(10, 6))
    plt.plot(y[0, :, 0], label='Actual')
    plt.plot(predictions[0, :, 0], label='Predicted')
    plt.legend()
    plt.show()

plot_predictions(feedback_model, val_X, val_y)
```
